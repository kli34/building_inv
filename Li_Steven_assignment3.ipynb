```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns
import bqplot
from ipywidgets import Layout
from ipywidgets import widgets
from IPython.display import display
```

```python
data = pd.read_csv('building_inventory.csv')
# data
```

# Relationship between the year acquired and the year constructed

```python
data1 = data.loc[(data['Year Acquired'] != 0) & (data['Year Constructed'] != 0)]

fig, ax = plt.subplots(figsize=(15,15))

ax.scatter(data1['Year Acquired'], data1['Year Constructed'])
ax.set_title('Relationship Between Year Acquired & Year Constructed', fontsize = 20)
ax.set_xlabel('Year Acquired', fontsize = 15)
ax.set_ylabel('Year Constructed', fontsize = 15)
# fig = px.scatter(x = data1['Year Acquired'], y = data1['Year Constructed'])
# fig.update_xaxes(title_text='Year Acquired')
# fig.update_yaxes(title_text='Year Constructed')
# fig.show()
```

I plotted a scatter plot to observe the relationship between year acquired and year constructed. I observed that there is some positive linear relationship between these two variables so many buildings have acquired and constructed at the same year. I also find out lots data have constructed then got acquired on the scatter plot.

# Total square footage as a function of congressional district ("Congress Dist")

```python
data2 = data.groupby('Congress Dist').agg({'Square Footage':'sum'}).reset_index()
# fig = px.bar(x = data2['Congress Dist'], y = data2['Square Footage'], color = data2['Congress Dist'], text = data2['Square Footage'])
# fig.update_xaxes(title_text='Congress Dist', dtick=1, tickangle=45, tickfont = dict(family='Rockwell', color='blue', size=12))
# fig.update_yaxes(title_text='Total Square Footage')
# fig.show()

fig, ax = plt.subplots(figsize=(15,15))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x='Congress Dist', y='Square Footage', data=data2)
ax.set_title('Total square footage by Congressinal District', fontsize = 20)
ax.set_xlabel('Congress Dist', fontsize = 15)
ax.set_ylabel('Total square footage', fontsize = 15)
```

On the bar plot, I can clearly see the total square footage by each congree district then District 13 actually has the highest total square footage.

# Average square footage per floor as a function of congressional district

```python
data3 = data.groupby('Congress Dist').agg({
    'Square Footage':'sum',
    'Total Floors':'sum'
}).reset_index()
data3['Square_Footage_per_floor'] = data3['Square Footage'] / data3['Total Floors']

# fig = px.bar(x = data3['Congress Dist'], y = data3['Square_Footage_per_floor'],
#             color = data3['Congress Dist'], text = data3['Square_Footage_per_floor'])
# fig.update_xaxes(title_text='Congress Dist', dtick=1, tickangle=45, tickfont = dict(family='Rockwell', color='blue', size=12))
# fig.update_yaxes(title_text='Square_Footage_per_floor')
# fig.show()

fig, ax = plt.subplots(figsize=(15,15))
sns.set_theme(style = "whitegrid")
ax = sns.barplot(x ='Congress Dist', y ='Square_Footage_per_floor', data = data3)
ax.set_title('Average square footage per floor by congressional district', fontsize = 20)
ax.set_xlabel('Congress Dist', fontsize = 15)
ax.set_ylabel('Total square footage', fontsize = 15)
```

I created bar plot again to visualize average square footage per floor by congression distrivt and most of districts have really high square footage per floor but the highest one goes to district 3.
